```python
#ex 1
def main():
    basketball_players = {}

    def add_player():
        fio = input("Введите ФИО баскетболиста: ")
        height = input("Введите рост баскетболиста: ")
        basketball_players[fio] = height
        print(f"Игрок {fio} добавлен.")

    def remove_player():
        fio = input("Введите ФИО баскетболиста для удаления: ")
        if fio in basketball_players:
            del basketball_players[fio]
            print(f"Игрок {fio} удален.")
        else:
            print("Игрок не найден.")

    def search_player():
        fio = input("Введите ФИО баскетболиста для поиска: ")
        if fio in basketball_players:
            print(f"Рост {fio}: {basketball_players[fio]}")
        else:
            print("Игрок не найден.")

    def replace_player():
        fio = input("Введите ФИО баскетболиста для замены: ")
        if fio in basketball_players:
            height = input("Введите новый рост баскетболиста: ")
            basketball_players[fio] = height
            print(f"Рост {fio} обновлен.")
        else:
            print("Игрок не найден.")

    while True:
        print("\nМеню:")
        print("1. Добавить баскетболиста")
        print("2. Удалить баскетболиста")
        print("3. Найти баскетболиста")
        print("4. Обновить рост баскетболиста")
        print("5. Выйти")

        choice = input("Выберите опцию: ")

        if choice == "1":
            add_player()
        elif choice == "2":
            remove_player()
        elif choice == "3":
            search_player()
        elif choice == "4":
            replace_player()
        elif choice == "5":
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

```python
#ex 2
def main():
    english_french_dict = {}

    def add_word():
        english_word = input("Введите слово на английском: ")
        french_translation = input("Введите перевод на французский: ")
        english_french_dict[english_word] = french_translation
        print(f"Слово {english_word} добавлено с переводом {french_translation}.")

    def remove_word():
        english_word = input("Введите слово для удаления на английском: ")
        if english_word in english_french_dict:
            del english_french_dict[english_word]
            print(f"Слово {english_word} удалено.")
        else:
            print("Слово не найдено.")

    def search_word():
        english_word = input("Введите слово для поиска на английском: ")
        if english_word in english_french_dict:
            print(f"Перевод на французский: {english_french_dict[english_word]}")
        else:
            print("Слово не найдено.")

    def replace_word():
        english_word = input("Введите слово для замены на английском: ")
        if english_word in english_french_dict:
            french_translation = input("Введите новый перевод на французский: ")
            english_french_dict[english_word] = french_translation
            print(f"Перевод для {english_word} обновлен на {french_translation}.")
        else:
            print("Слово не найдено.")

    while True:
        print("\nМеню:")
        print("1. Добавить слово")
        print("2. Удалить слово")
        print("3. Найти слово")
        print("4. Обновить перевод")
        print("5. Выйти")

        choice = input("Выберите опцию: ")

        if choice == "1":
            add_word()
        elif choice == "2":
            remove_word()
        elif choice == "3":
            search_word()
        elif choice == "4":
            replace_word()
        elif choice == "5":
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
