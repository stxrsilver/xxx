```python
# ex 1
def main():
    ids = [101, 203, 150]
    phone_numbers = ["+123456789", "+987654321", "+567890123"]

    while True:
        print("\nМеню:")
        print("1. Отсортировать по идентификационным кодам")
        print("2. Отсортировать по номерам телефонов")
        print("3. Вывести список пользователей с кодами и телефонами")
        print("4. Выход")
        choice = input("Выберите пункт меню: ")

        if choice == "1":
            if not ids:
                print("Список пуст. Добавьте данные для сортировки.")
            else:
                sorted_data = sorted(zip(ids, phone_numbers), key=lambda x: x[0])
                ids, phone_numbers = zip(*sorted_data)
                ids, phone_numbers = list(ids), list(phone_numbers)
                print("\nСписок отсортирован по идентификационным кодам:")
                for i, phone in zip(ids, phone_numbers):
                    print(f"Код: {i}, Телефон: {phone}")
        elif choice == "2":
            if not phone_numbers:
                print("Список пуст. Добавьте данные для сортировки.")
            else:
                sorted_data = sorted(zip(ids, phone_numbers), key=lambda x: x[1])
                ids, phone_numbers = zip(*sorted_data)
                ids, phone_numbers = list(ids), list(phone_numbers)
                print("\nСписок отсортирован по номерам телефонов:")
                for i, phone in zip(ids, phone_numbers):
                    print(f"Код: {i}, Телефон: {phone}")
        elif choice == "3":
            if not ids:
                print("Список пуст.")
            else:
                print("\nСписок пользователей:")
                for i, phone in zip(ids, phone_numbers):
                    print(f"Код: {i}, Телефон: {phone}")
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
