```python
#ex 1
def linear_search(lst, target):
    for index, value in enumerate(lst):
        if value == target:
            return index
    return -1

def main():
    list1 = [5, 12, 7]
    list2 = [3, 8, 10]
    list3 = [1, 4, 6]
    list4 = [9, 2, 11]

    combined_list = list1 + list2 + list3 + list4

    while True:
        print("\nМеню:")
        print("1. Отсортировать по возрастанию")
        print("2. Отсортировать по убыванию")
        print("3. Найти значение")
        print("4. Выход")
        choice = input("Выберите пункт меню: ")

        if choice == "1":
            combined_list.sort()
            print("\nСписок отсортирован по возрастанию:")
            print(combined_list)
        elif choice == "2":
            combined_list.sort(reverse=True)
            print("\nСписок отсортирован по убыванию:")
            print(combined_list)
        elif choice == "3":
            try:
                target = int(input("Введите значение для поиска: "))
                index = linear_search(combined_list, target)
                if index != -1:
                    print(f"Значение {target} найдено на позиции {index}.")
                else:
                    print(f"Значение {target} не найдено.")
            except ValueError:
                print("Ошибка: введите целое число.")
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

```python
#ex 2
def binary_search(lst, target):
    left, right = 0, len(lst) - 1
    while left <= right:
        mid = (left + right) // 2
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

def main():
    list1 = [1, 2, 3, 4]
    list2 = [3, 4, 5, 6]
    list3 = [6, 7, 8, 9]
    list4 = [9, 10, 11, 12]

    unique_elements = set(list1) ^ set(list2) ^ set(list3) ^ set(list4)
    unique_list = list(unique_elements)

    while True:
        print("\nМеню:")
        print("1. Отсортировать по возрастанию")
        print("2. Отсортировать по убыванию")
        print("3. Найти значение")
        print("4. Выход")
        choice = input("Выберите пункт меню: ")

        if choice == "1":
            unique_list.sort()
            print("\nСписок отсортирован по возрастанию:")
            print(unique_list)
        elif choice == "2":
            unique_list.sort(reverse=True)
            print("\nСписок отсортирован по убыванию:")
            print(unique_list)
        elif choice == "3":
            if not unique_list:
                print("Список пуст. Добавьте данные для поиска.")
            else:
                try:
                    target = int(input("Введите значение для поиска: "))
                    unique_list.sort()
                    index = binary_search(unique_list, target)
                    if index != -1:
                        print(f"Значение {target} найдено на позиции {index}.")
                    else:
                        print(f"Значение {target} не найдено.")
                except ValueError:
                    print("Ошибка: введите целое число.")
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
